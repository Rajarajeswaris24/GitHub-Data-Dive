collecting repositories

```python
import requests

headers = {"Authorization": "your_access_token"}
topics = ['Machine Learning',  'Deep Learning', 'NLP','Data Visualization', 'Artificial Intelligence','Big Data','SQL','Cloud Data Platforms','MongoDB','Statistics']

repositories = []

for topic in topics:
    response = requests.get(
        f"https://api.github.com/search/repositories?q=topic:{topic}&sort=stars",
        headers=headers
    )
    data = response.json()['items']
    for repo in data:
        repositories.append({
            'Repository_Name': repo['name'],
            #'topic': topic,
            'Owner': repo['owner']['login'],
            'Description': repo['description'],
            'URL': repo['html_url'],
            'Programming_Language': repo['language'],
            'Creation_Date': repo['created_at'],
            'Last_Updated_Date': repo['updated_at'],
            'Number_of_Stars': repo['stargazers_count'],
            'Number_of_Forks': repo['forks_count'],
            'Number_of_Open_Issues': repo['open_issues_count'],
            'License_Type': repo['license']['name'] if repo['license'] else 'Not Specified'
        })
```

```python
cont_col=df2.select_dtypes(exclude=['object','datetime64[ns]']).columns
cont_col
```

```python

for column in cont_col:

    print(f"Column : {column}")

    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    upper_threshold = df[column].quantile(0.75) + (1.5 * iqr)
    lower_threshold = df[column].quantile(0.25) - (1.5 * iqr)
    print(f"UpperThreshold : {upper_threshold} and lowerThreshold : {lower_threshold}")
    print(df[column].describe())

    print('**********************')
```

```python


for column in cont_col:

    print(f"Column : {column}")

    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    upper_threshold = df[column].quantile(0.75) + (1.5 * iqr)
    lower_threshold = df[column].quantile(0.25) - (1.5 * iqr)
    print(f"UpperThreshold : {upper_threshold} and lowerThreshold : {lower_threshold}")

    df[column] = df[column].clip(lower_threshold, upper_threshold)   #cliping- removing outliers
    print(df[column].describe())


    print('**********************')
```
